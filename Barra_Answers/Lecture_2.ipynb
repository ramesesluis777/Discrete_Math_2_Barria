$\textbf{Consider the following experiment:}$

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

$\textbf{Question 1:}$ Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

$\textit{Proof:}$

To prove that the expected value $ E[y] = \sum y \cdot P(y) = n \cdot p_i $, we start with the definition of expected value:

$ E[y] = \sum y \cdot P(y) $

Given the joint probability distribution $ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $, where $ y_i $ represents the number of trials for which the outcome falls into cell $ i $, and $ P_i $ represents the probability that the outcome falls into cell $ i $, we have:

$ E[y] = \sum y \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $

Now, let's rewrite the expected value in terms of the individual probabilities $ p_i $:

$ E[y] = \sum (y_1 \cdot P_1 + y_2 \cdot P_2 + y_3 \cdot P_3 + \dotsb + y_n \cdot P_n) \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} $

$ = \sum \left( y_1 \cdot \frac{n!}{y_1!} \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} + \dotsb + y_n \cdot \frac{n!}{y_n!} \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} \right) $

$ = \sum \left( n \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} \right) $

$ = n \cdot \left( P_1^1 + P_2^2 + P_3^3 + \dotsb + P_n^n \right) $

$ = n \cdot (p_1 + p_2 + p_3 + \dotsb + p_n) $

Since $ p_1 + p_2 + p_3 + \dotsb + p_n = 1 $ (as stated in the problem), we have:

$ E[y] = n \cdot 1 = n \cdot p_i $

This proves that the expected value $ E[y] = \sum y \cdot P(y) = n \cdot p_i $.


$\textbf{Question 2:}$ Given the following data:

\begin{bmatrix}
\begin{array}{c|c}
Age & Proportion \\
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\end{bmatrix}

If 500 adults are sampled  randomly, find the probability that the sample contains 100 persons between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?

$\textit{Proof:}$

To find the probability that the sample contains 100 persons between the ages of 18 and 24, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64, we'll use the multinomial probability formula.

Given:
- $n = 500$ (total number of adults sampled)
- $k = 5$ (number of age groups)
- $p_1 = 0.18$, $p_2 = 0.23$, $p_3 = 0.16$, $p_4 = 0.27$, $p_5 = 0.16$ (proportions for each age group)

The probability mass function (PMF) for a multinomial distribution is:

$P(y) = \frac{n!}{y_1!y_2!y_3!y_4!y_5!} p_1^{y_1} p_2^{y_2} p_3^{y_3} p_4^{y_4} p_5^{y_5}$

Where:
- $y_1$, $y_2$, $y_3$, $y_4$, $y_5$ are the counts for each age group

For the given scenario, we have:
- $y_1 = 100$ (for ages 18-24)
- $y_2 = 200$ (for ages 25-34)
- $y_3 = 0$ (no persons between ages 35-44)
- $y_4 = 200$ (for ages 45-64)
- $y_5 = 500 - (y_1 + y_2 + y_3 + y_4) = 500 - (100 + 200 + 0 + 200) = 0$ (for ages 65-100)

Substitute these values into the PMF and calculate:

$P(y) = \frac{500!}{100!200!0!200!0!} (0.18)^{100} (0.23)^{200} (0.16)^{0} (0.27)^{200} (0.16)^{0}$

$P(y) = \frac{500!}{100!200!200!} (0.18)^{100} (0.23)^{200} (0.27)^{200}$

The expected value to obtain a person in the 65 and above age group is given by the sum of the products of each outcome and its probability:

$E[Y_5] = n \times p_5 = 500 \times 0.16 = 80$

So, the expected number of persons aged 65 and above is 80.

$\textbf{Consider the following experiment:}$

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

$\textbf{Question 1:}$ From the following steps shown above, derive the probability distribution of the experiment.

$\textit{Proof:}$

To derive the probability distribution of the experiment, we use the binomial distribution.

The PMF of a binomial distribution is given by:

$P(Y = k) = \binom{n}{k} p^k q^{n-k}$

Where:
- $k$ is the number of successes,
- $n$ is the total number of trials,
- $p$ is the probability of success on a single trial,
- $q = (1 - p)$ is the probability of failure on a single trial,
- $\binom{n}{k}$ is the binomial coefficient.

Substituting the given values, we get:

$P(Y = k) = \binom{n}{k} p^k (1 - p)^{n-k}$

This is the probability distribution of the experiment, where $k$ can range from 0 to $n$.


$\textbf{Question 2:}$ Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.

$\textit{Proof:}$

To find the expectation of the probability distribution, $E[Y]$, we need to compute the sum of each possible outcome $y$ multiplied by its corresponding probability $P(y)$.

For a binomial distribution with parameters $n$ and $p$, where $n$ is the number of trials and $p$ is the probability of success on a single trial, the expectation $E[Y]$ is given by:

$E[Y] = \sum_{k=0}^{n} y \cdot P(Y = y)$

Substituting the expression for $P(Y = y)$ from the previous derivation:

$E[Y] = \sum_{k=0}^{n} y \cdot \binom{n}{k} p^k (1 - p)^{n-k}$

Now, for each term in the summation, $y$ is equal to $k$, as $y$ is the number of successes, which is $k$, and the probability $P(Y = y)$ is $\binom{n}{k} p^k (1 - p)^{n-k}$.

$$
\begin{align*}
E[Y] & = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n-k} \\
& = \sum_{k=0}^{n} k \cdot \frac{n!}{k!(n-k)!} p^k (1 - p)^{n-k}
\end{align*}
$$

Now, let's rewrite the sum in terms of the binomial expansion:

$E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n-k}$

$E[Y] = np\sum_{k=0}^{n} \binom{n-1}{k-1} p^{k-1} (1 - p)^{n-k}$

Now, the summation is a binomial expansion of $(p + (1 - p))^n = 1^n = 1$. Therefore, the sum becomes 1.

$E[Y] = np$

This proves that the expectation of the binomial distribution is $E[Y] = np$.


$\textbf{Question 3:}$ Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

$\textit{Proof:}$

To solve this problem, we can use the binomial distribution formula. Let $p = 0.30$ be the probability of recovery for a single individual and $q = 1 - p = 0.70$ be the probability of failure to recover for a single individual. We want to find the probability of at least nine out of ten individuals recovering.

The probability mass function (PMF) of the binomial distribution is given by:

$P(Y = k) = \binom{n}{k} p^k q^{n-k}$

where $n$ is the total number of trials (number of individuals), $k$ is the number of successes (number of individuals recovering), and $\binom{n}{k}$ is the binomial coefficient.

For this problem, $n = 10$, $p = 0.30$, and $q = 0.70$. We want to find $P(Y \geq 9)$.

Now, let's calculate the probability using the formula:

$$
\begin{align*}
P(Y \geq 9) & = P(Y = 9) + P(Y = 10) \\
& = \binom{10}{9} p^9 q^{10-9} + \binom{10}{10} p^{10} q^{10-10}\\
& = \binom{10}{9} (0.30)^9 (0.70)^{1} + \binom{10}{10} (0.30)^{10}(1)\\
& = 10 \times (0.30)^9 \times (0.70)^{1} + 1 \times (0.30)^{10} \\
& \approx 0.0001377 + 0.0000059 \\
& \approx 0.000144
\end{align*}
$$

So, the probability that at least nine out of ten individuals receiving the medication will recover is approximately 0.000144.


$\textbf{Question 4:}$ Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

$\textit{Proof:}$
To find the probability of observing at least one defective fuse out of a sample of 5, given that 5% of the fuses are defective, we can use the complement rule. The complement rule states that the probability of an event occurring is equal to 1 minus the probability of the event not occurring.

Let's denote:
- $ p $ = probability of selecting a defective fuse = 0.05
- $ q $ = probability of selecting a non-defective fuse = 1 - 0.05 = 0.95
- $ n $ = number of trials (number of fuses selected) = 5
- $ Y $ = number of successes (number of defective fuses observed)

We want to find $ P(Y \geq 1) $, the probability of observing at least one defective fuse.

Using the complement rule, we have:

$ P(Y \geq 1) = 1 - P(Y = 0) $

Now, the probability of observing 0 defective fuses out of 5 can be calculated using the binomial probability formula:

$ P(Y = k) = \binom{n}{k} \times p^k \times q^{(n-k)} $

where:
- $ n $ = number of trials (5)
- $ k $ = number of successes (0 in this case)
- $ p $ = probability of success (0.05)
- $ q $ = probability of failure (0.95)

So,

$ P(Y = 0) = \binom{5}{0} \times 0.05^0 \times 0.95^5 $

$ = 1 \times 1 \times (0.95)^5 $

Now, we can plug this into the complement rule:

$ P(Y \geq 1) = 1 - (0.95)^5 $

Let's calculate this probability:

$ P(Y \geq 1) = 1 - (0.95)^5 \approx 1 - 0.7737809375 \approx 0.2262190625 $

So, the probability of observing at least one defective fuse in a sample of 5 is approximately $0.2262$, or $22.62\%$.

Consider a probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.


$\textbf{Question 1:}$ Find a general formula for the expected value of this distribution.

$\textit{Proof:}$

To find the expected value of this distribution, denoted as $E[Y]$, we use the formula:

$$E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$$

Substituting the given probability distribution, we get:

$$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$

Now, let's simplify this expression:

$$
\begin{align*}
E[Y] & = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} \\
& = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} \\
& = e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} \\
& = e^{-\lambda} \lambda \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} \\
& = e^{-\lambda} \lambda \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} \\
& = e^{-\lambda} \lambda e^{\lambda} \quad \text{(Using the Taylor series expansion of } e^x \text{)} \\
& = \lambda
\end{align*}
$$

Therefore, the general formula for the expected value of this distribution is $E[Y] = \lambda$.


$\textbf{Question 2:}$ A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

$\textit{Proof:}$

To solve this problem, we can use the Poisson distribution, as the mean density of seedlings per square yard follows a Poisson distribution with parameter $\lambda = 5$.

The probability mass function (PMF) of the Poisson distribution is given by:

$$P(Y = y) = \frac{e^{-\lambda} \lambda^y}{y!}$$

where:
- $Y$ is the number of seedlings in a 1-square-yard region,
- $\lambda$ is the mean number of seedlings per square yard,
- $e$ is the base of the natural logarithm, approximately equal to 2.71828.

For this problem, $\lambda = 5$ and $y = 0$, as we want to find the probability that none of the regions will contain seedlings.

Now, let's calculate the probability using the Poisson distribution formula:

$$P(Y = 0) = \frac{e^{-5} 5^0}{0!}$$

$$P(Y = 0) = e^{-5}$$

$$P(Y = 0) \approx 0.0067$$

Therefore, the probability that none of the regions will contain seedlings is approximately 0.0067.
